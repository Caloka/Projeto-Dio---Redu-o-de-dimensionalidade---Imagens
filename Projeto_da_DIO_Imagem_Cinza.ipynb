### Projeto da DIO - Redução de Dimensionalidade em Imagens
#### Resumo: Será transformado uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255).

```python
import cv2
import numpy as np
```

```python
imagem_colorida = r"C:\Users\Usuario\Downloads\img_color.jpg"
ler_imagem = cv2.imread(imagem_colorida)
ler_imagem = cv2.cvtColor(ler_imagem, cv2.COLOR_BGR2GRAY)
suavização = cv2.GaussianBlur(ler_imagem, (7,7), 0)
(T, bin) = cv2.threshold(suavização, 160, 255, cv2.THRESH_BINARY)
(T, binI) = cv2.threshold(suavização, 160, 255, cv2.THRESH_BINARY_INV)
resultado = np.vstack([
  np.hstack([suavização,bin]),
  np.hstack([binI, cv2.bitwise_and(ler_imagem, ler_imagem, mask=binI)])
])

cv2.imshow("Binarização da imagem", resultado)
cv2.waitKey(0)
```
